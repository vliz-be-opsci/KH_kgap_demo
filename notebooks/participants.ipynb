#### Publications of the KH participants

```python
participant_publications = execute_to_df('participants-publications.sparql')
participant_publications['name'] =  participant_publications['first_name'] + ' ' + participant_publications['last_name'] 

#explore
participant_publications
```

```python
partcipant_pub_count = participant_publications.groupby('person').nunique()
partcipant_pub_count
```

```python
affiliation_pub_count = participant_publications.groupby('affiliation_name').nunique()
affiliation_pub_count
```

```python
affiliation_pub_count = participant_publications.groupby('affiliation').nunique()
affiliation_pub_count
```

```python
participant_pub_count = participant_publications.groupby('name').nunique()
participant_pub_count.drop(columns=['first_name','last_name','person'], inplace=True)
participant_pub_count.reset_index(inplace=True)
```

```python
participant_pub_count.plot(kind='bar', x='name', y='pub', color='skyblue')
plt.show()
```

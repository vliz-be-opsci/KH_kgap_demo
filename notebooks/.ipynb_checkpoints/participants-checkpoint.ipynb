#### 3. Information of the number of publications of the KH participants, grouped across institutes

In the next, we'll have a look at the publication that are associated with people (via their OrcID), while also listing their affiliation.  
This allows us to analyse how many publications are linked to each affiliation.

```python
participant_pubs = execute_to_df('participants_pubs.sparql')

participant_pubs['name'] = participant_pubs['first_name'] + ' ' + participant_pubs['last_name']
participant_pubs
```

```python
# Number of publications from registered participants, grouped per institute with which each registered participant is currently associated
affiliation_pub_count = participant_pubs.groupby('current_affiliation').nunique()
affiliation_pub_count.drop(columns=['person','first_name','last_name','name'], inplace=True)
affiliation_pub_count.reset_index(inplace=True)
# table (you see the importance of using unique and persistent identifiers for things (people, institutes, ...) )
affiliation_pub_count
```

```python
# graph
affiliation_pub_count.plot(kind='bar', x='current_affiliation', y='pub', color='skyblue')
plt.show()
```

```python
# Number of publications from registered participants, grouped per registered participant
participant_pub_count = participant_pubs.groupby('name').nunique()
participant_pub_count.drop(columns=['first_name','last_name','person', 'current_affiliation'], inplace=True)
participant_pub_count.reset_index(inplace=True)

#table
participant_pub_count
```

```python
#graph 
participant_pub_count.plot(kind='bar', x='name', y='pub', color='skyblue')
plt.show()
```
